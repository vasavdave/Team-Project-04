```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
diabetes_df = pd.read_csv(Path('diabetes-dataset.csv'))
pimaData_df = pd.read_csv(Path('diabetes_PIMA_INDIANS.csv'))
diabetes_df.head()
```

```python
#df_new = df.drop(df[(df['col_1'] == 1.0) & (df['col_2'] == 0.0)].index)
diabetes_df_cl = diabetes_df.drop(diabetes_df[(diabetes_df['BloodPressure']==0) | (diabetes_df['Glucose']==0)
                    | (diabetes_df['SkinThickness']==0) | (diabetes_df['Insulin']==0)
                    | (diabetes_df['Age']==0) | (diabetes_df['BMI']==0)].index)

diabetes_df_cl.describe()
```

```python
pimaData_df_cl = pimaData_df.drop(pimaData_df[(pimaData_df['BloodPressure']==0) | (pimaData_df['Glucose']==0)
                    | (pimaData_df['SkinThickness']==0) | (pimaData_df['Insulin']==0)
                    | (pimaData_df['Age']==0) | (pimaData_df['BMI']==0)].index)

pimaData_df_cl.describe()
```

```python
pimaData_df_cl.to_csv("CleanerPimaData.csv", index=False)
```

```python
diabetes_df_cl.to_csv("CleanerDiabetesData.csv", index=False)
```

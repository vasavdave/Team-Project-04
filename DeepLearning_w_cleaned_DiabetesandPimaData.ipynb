```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
import sklearn as skl
```

```python
diabetes_df = pd.read_csv(Path('CleanerDiabetesData.csv'))
pimaData_df = pd.read_csv(Path('CleanerPimaData.csv'))
diabetes_df.head()
```

```python
factors_df = diabetes_df.drop('Outcome', axis = 1)
#Ind_df = pd.get_dummies(Independent_df)
factors_df.head()
```

```python
diab_out_df = diabetes_df.Outcome
diab_out_df.head()
```

```python
X = factors_df
y = diab_out_df

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=18)
```

```python
from sklearn.preprocessing import StandardScaler
X_scaler = skl.preprocessing.StandardScaler()
X_scaler.fit(X_train)
#scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
import tensorflow as tf
```

```python
nn_model = tf.keras.models.Sequential()
nn_model.add(tf.keras.layers.Dense(units=25, activation="relu", input_dim=8))
nn_model.add(tf.keras.layers.Dense(units=25, activation="relu"))
nn_model.add(tf.keras.layers.Dense(units=25, activation="relu"))
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
nn_model.summary()
```

```python
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Fit the model to the training data
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=100)
```

```python
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# use model on Pima Indian data:
PimaFactors_df = diabetes_df.drop('Outcome', axis = 1)
#Ind_df = pd.get_dummies(Independent_df)
PimaFactors_df.head()
```

```python
Pima_diab_out_df = diabetes_df.Outcome
Pima_diab_out_df.head()
```

```python
X_Pima_scaled = X_scaler.transform(PimaFactors_df)
```

```python
model_loss, model_accuracy = nn_model.evaluate(X_Pima_scaled,Pima_diab_out_df,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
